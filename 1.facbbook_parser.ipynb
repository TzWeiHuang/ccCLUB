```python
totle_posts = []
totle_messages = []
totle_time = []
n = 0
t = 0

while len(totle_posts) < 50:
    more_messages = driver.find_elements(By.XPATH, '/html/body/div/div/div[2]/div/div[1]/div[3]/section/article[position()>=1]/footer/div[2]/a[2]')
    for i in more_messages:
# open post in new page
        try:
            new_web = i.get_attribute('href')
            driver.execute_script("window.open('');")
            driver.switch_to.window(driver.window_handles[-1])
            driver.get(new_web)
        except:
            continue
# post
        try:
            post = driver.find_element(By.XPATH, '/html/body/div/div/div[2]/div/div[2]/div[1]/div/div/div/div/p')
            if "鼻" in post.text:
                totle_posts.append(post.text)
                n+=1
                t+=1
                print(n)
            else:
                t+=1
                driver.close()
                driver.switch_to.window(driver.window_handles[0])
                continue
        except:
            try:
                post = driver.find_element(By.XPATH, '/html/body/div/div/div[2]/div/div[2]/div[1]/div/div/div/div/div/div/div/span')
                if "鼻" in post.text:
                    totle_posts.append(post.text)
                    n+=1
                    t+=1
                    print(n)
                else:
                    t+=1
                    driver.close()
                    driver.switch_to.window(driver.window_handles[0])
                    continue
            except:
                driver.close()
                driver.switch_to.window(driver.window_handles[0])
                continue
# time 
        time = driver.find_element(By.XPATH, '/html/body/div/div/div[2]/div/div[2]/div[1]/div/footer/div/abbr')
        totle_time.append(time.text)
# messages
        messages_lst= []
        while True:
            try:
                messages = driver.find_elements(By.XPATH, '/html/body/div/div/div[2]/div/div[2]/div[2]/div/div[3]/div[position()>=1]/div/div[1]')
                for i in messages:
                    messages_lst.append(i.text)
# show_before
                show_before = driver.find_element(By.XPATH,'/html/body/div/div/div[2]/div/div[2]/div[2]/div/div[3]/div[1]/a')
                show_before.click()

            except: 
                driver.close()
                driver.switch_to.window(driver.window_handles[0])
                break


        totle_messages.append(messages_lst)

    
# open more post
    more_posts = driver.find_element(By.XPATH, '/html/body/div/div/div[2]/div/div[1]/div[3]/div')
    more_posts.click()
# pandas
dict_data_frame = {"post":totle_posts, "message":totle_messages, "time":totle_time}
df = pd.DataFrame(dict_data_frame)
print("total=",t)
print(df)
df.to_csv('ccClub_project_facebook.csv')
```
