```python
from ckiptagger import WS, POS, NER
import pandas as pd
df = pd.read_csv("ccClub_project_facebook.csv")   
```

```python
# find names in messages
total_ner_result = []
n=0
for i in range(51):
    messages = df.iloc[i,2]
    text = messages
    ws = WS("./data")
    pos = POS("./data")
    ner = NER("./data")

    ws_results = ws([text])
    pos_results = pos(ws_results)
    ner_results = ner(ws_results, pos_results)

    total_ner_result.append(ner_results)
    
    n += 1
    print(n)
    for name in ner_results[0]:
        print(name)

    print("=============")
df["name"]=total_ner_result
print(df)
```

```python
df.to_csv('ccClub_project_facebook.csv')
```
